```python
import plotly.express as px
import pandas as pd

# Data setup
data = {
    'Category': ['Statistical', 'Math & Trig', 'Financial', 'Compatibility', 'Engineering', 
                 'Lookup & Reference', 'Text', 'Date & Time', 'Information', 'Database', 'Logical', 'Web'],
    'Function Count': [102, 75, 54, 51, 40, 31, 28, 25, 20, 15, 11, 5]
}
df = pd.DataFrame(data)

# 1. Bar Chart with Plotly Express
fig1 = px.bar(df, x='Category', y='Function Count', 
             title='Excel Functions Count by Category',
             color='Function Count',
             color_continuous_scale='Viridis',
             text='Function Count')
fig1.show()

# 2. Pie Chart with Plotly Express
fig2 = px.pie(df, values='Function Count', names='Category', 
             title='Distribution of Excel Functions',
             hole=0.4)
fig2.show()
```
